# Exercici 4
Mostreu-me amb exemples de diferents matrius, la regla fonamental de Broadcasting que diu : "les matrius es poden transmetre / broadcast si les seves dimensions coincideixen o si una de les matrius té una mida d'1".

```python
# Create an array 

array_8 = np.array([r.randint(0,10) for i in range(8)])
print(array_8)

array_1 = np.array([r.randint(0,10) for i in range(1)])
print(array_1)

array_2 = np.array([r.randint(0,10) for i in range(2)])
print(array_2)

array_3 = np.array([r.randint(0,10) for i in range(3)])
print(array_3)
```

```python
# Operating between them: fundamental rule "una de les matrius té una mida d'1"

print("array_8 * array_2 = ",array_8 * array_1) # the value of array_2 with shape 1, is multiplied by all elements of array_1
print("array_8 + array_2 = ",array_8 + array_1) # the value of array_2 with shape 1, is added by all elements of array_1
```

```python
# Error is retrieve when operating with different array's shapes
print(array_8 + array_2)
```

```python
# Operating between them: fundamental rule "si les seves dimensions coincideixen"

array_8x3_a = np.array([[r.randint(0,10) for i in range(8)],
                      [r.randint(10,20) for i in range(8)],
                      [r.randint(20,30) for i in range(8)]])
array_8x3_b = np.array([[r.randint(0,1) for i in range(8)],
                      [r.randint(2,3) for i in range(8)],
                      [r.randint(3,4) for i in range(8)]])
print(array_8x3_a)
print(array_8x3_b)
```

```python
print("array_8x3_a - array_8x3_b = \n",
      array_8x3_a - array_8x3_b , "\n") # Array 8x3 (3D) - Array 8x3 (3D)

print("array_8x3_a - array_8 = \n",
      array_8x3_a - array_1) # Array 8x3 (3D) - Array 8x1 (1D)
```
